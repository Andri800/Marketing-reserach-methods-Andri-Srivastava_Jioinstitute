```
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re

# Load the dataset
file_path = "/content/Mobiles Dataset (2025).csv"  # Replace with your actual file path in Colab
df = pd.read_csv(file_path, encoding="latin1")

# Convert 'Launched Year' to integer
df['Launched Year'] = pd.to_numeric(df['Launched Year'], errors='coerce')

# Function to extract numeric values from text columns
def extract_numeric(value):
    if isinstance(value, str):
        num = re.findall(r'\d+', value.replace(',', ''))  # Extract numbers
        return int(num[0]) if num else None
    return value

# Convert relevant columns to numeric
df['RAM (GB)'] = df['RAM'].apply(extract_numeric)
df['Battery Capacity (mAh)'] = df['Battery Capacity'].apply(extract_numeric)
df['Front Camera (MP)'] = df['Front Camera'].apply(extract_numeric)
df['Back Camera (MP)'] = df['Back Camera'].apply(extract_numeric)

# Extract numeric price (USD)
df['Launched Price (USA)'] = df['Launched Price (USA)'].apply(extract_numeric)

# Group data by 'Launched Year' and calculate average specs
trend_data = df.groupby('Launched Year').agg({
    'Launched Price (USA)': 'mean',
    'RAM (GB)': 'mean',
    'Battery Capacity (mAh)': 'mean',
    'Front Camera (MP)': 'mean',
    'Back Camera (MP)': 'mean'
}).reset_index()

# Plot trends
plt.figure(figsize=(12, 6))
sns.lineplot(x='Launched Year', y='Launched Price (USA)', data=trend_data, marker='o', label="Avg. Price (USD)")
sns.lineplot(x='Launched Year', y='RAM (GB)', data=trend_data, marker='s', label="Avg. RAM (GB)")
sns.lineplot(x='Launched Year', y='Battery Capacity (mAh)', data=trend_data, marker='^', label="Avg. Battery (mAh)")
sns.lineplot(x='Launched Year', y='Front Camera (MP)', data=trend_data, marker='d', label="Avg. Front Camera (MP)")
sns.lineplot(x='Launched Year', y='Back Camera (MP)', data=trend_data, marker='v', label="Avg. Back Camera (MP)")

plt.title("Evolution of Smartphone Specifications Over the Years")
plt.xlabel("Year")
plt.ylabel("Average Value")
plt.legend()
plt.grid(True)
plt.show()
```

```
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re

# Load the dataset
file_path = "/content/Mobiles Dataset (2025).csv"  # Replace with your actual file path in Colab
df = pd.read_csv(file_path, encoding="latin1")

# Function to extract numeric values from text columns
def extract_numeric(value):
    if isinstance(value, str):
        num = re.findall(r'\d+', value.replace(',', ''))  # Extract numbers
        return int(num[0]) if num else None
    return value

# Convert price columns to numeric
price_columns = ['Launched Price (Pakistan)', 'Launched Price (India)', 'Launched Price (China)',
                 'Launched Price (USA)', 'Launched Price (Dubai)']

for col in price_columns:
    df[col] = df[col].apply(extract_numeric)

# Convert 'Launched Year' to integer
df['Launched Year'] = pd.to_numeric(df['Launched Year'], errors='coerce')

# ---- 1. Compare Smartphone Prices Across Regions ----
avg_prices = df[price_columns].mean()

plt.figure(figsize=(10, 6))
sns.barplot(x=avg_prices.index, y=avg_prices.values, palette="coolwarm")
plt.xticks(rotation=45)
plt.title("Average Smartphone Prices in Different Regions")
plt.xlabel("Region")
plt.ylabel("Average Price (in local currency)")
plt.grid(axis="y")
plt.show()

# ---- 2. Identify Most & Least Expensive Brands/Models ----
most_expensive = df.loc[df['Launched Price (USA)'].idxmax(), ['Company Name', 'Model Name', 'Launched Price (USA)']]
least_expensive = df.loc[df['Launched Price (USA)'].idxmin(), ['Company Name', 'Model Name', 'Launched Price (USA)']]

print("Most Expensive Smartphone:")
print(most_expensive)
print("\nLeast Expensive Smartphone:")
print(least_expensive)

# ---- 3. Price Trends Over Launch Years ----
trend_data = df.groupby('Launched Year')[price_columns].mean().reset_index()

plt.figure(figsize=(12, 6))
for col in price_columns:
    sns.lineplot(x='Launched Year', y=col, data=trend_data, marker='o', label=col)

plt.title("Smartphone Price Trends Over Years")
plt.xlabel("Year")
plt.ylabel("Average Price")
plt.legend()
plt.grid(True)
plt.show()
```
